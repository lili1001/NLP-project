```python
import os
import json
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

import the corpus (raw data)

```python
processedDircPath=r'C:\Users\admin\OneDrive\Documents\supervised project\processedFiles'
rawDataFile=open(os.path.join(processedDircPath,'rawDataList.txt'), "r" )
rawDataList=json.loads(rawDataFile.read())
```

build function

```python
def initialize_count():
    # Initialize a CountVectorizer object: count_vectorizer
    countVec = CountVectorizer(stop_words="english", analyzer='word', 
                            ngram_range=(1, 1), max_df=1.0, min_df=1, max_features=20)
    countMatrix = countVec.fit_transform(rawDataList)
    # round decimal in countMatrix to 2
    countMatrixR=np.round(countMatrix , decimals=2).toarray() 
    return [countVec, countMatrixR]
```

```python
def initialize_TFIDF():
    # Initialize a TfidfVectorizer object: tfidf_vectorizer
    tfidfVec = TfidfVectorizer(stop_words="english", analyzer='word',
                                ngram_range=(1, 1), max_df=1.0, min_df=1, max_features=10)
    tfidfMatrix = tfidfVec.fit_transform(rawDataList)
    # round decimal in tfidfMatrix to 2
    tfidfMatrixR=np.round(tfidfMatrix, decimals=2).toarray() 
    return [tfidfVec, tfidfMatrixR]
```

```python
def matrix_present(vec,matrix):
    return pd.DataFrame(matrix, columns=vec.vocabulary_)
```

```python
def rank_present(vec, MatrixR):
    #rank key words for each doc based on TFIDF or TF
    sortedDF={}
    df=pd.DataFrame(MatrixR, columns=vec.vocabulary_).to_dict('index')
    for key, value in df.items():
        sortedValue=sorted(value.items(), key=lambda x: x[1], reverse=True)
        sortedDF[key]=sortedValue
    return pd.DataFrame.from_dict(sortedDF, orient='index')
```
